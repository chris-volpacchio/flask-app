```python
# Reading in images and measuring RGB pixels
import numpy as np, matplotlib.pyplot as plt, pandas as pd
from PIL import Image

img = './pics/test2.jpg'
img = Image.open(img, mode = 'r')
img = np.array(img)


def rgb_array(np_array):
    arrays = []
    for i in np_array:
        for a in i:
            # Why saving first 3 elems? rgb in 3+ elem vector? What other elems represent?
            a = list(a[:3])
            arrays.append(a)
    return pd.DataFrame(arrays, columns = ['r','g','b'])

df = rgb_array(img)

#print(df['r'].value_counts())
print(df)

# img = np.array(img)
# img[:1]

# set(np.bincount(img.ravel()))



# import imageio

# %matplotlib inline

# pic = imageio.imread('./test_img.png')
# plt.figure(figsize = (5,5))
# plt.imshow(pic);
```
